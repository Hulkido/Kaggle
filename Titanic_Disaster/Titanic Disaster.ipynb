```
def sigmoid(D):
    for i in range(D.shape[0]):
        for j in range(D.shape[1]):
            D[i,j]=(1/(1+math.exp(-D[i,j])))
    return D
```

```
def cost_function(theta,X,Y,lamb):
    product=np.dot(X,theta)
    h=sigmoid(product)
    part1=(-Y*np.log(h)-(1-Y)*np.log(h)/(np.alen(h))).sum()
    part2=((theta*theta[1:])*lamb/(2*np.alen(h))).sum(skipna=True)
    return part1+part2
```

```
def gradient_descent(X,Y,iter,lamb,alpha):
    initial_theta=np.zeros((np.shape(X)[1],1))
    for i in range(iter):
        h=sigmoid(np.dot(X,initial_theta))
        initial_theta=initial_theta-alpha*(X.transpose().dot(h-Y)+lamb*initial_theta)/np.alen(X)
    return initial_theta
```

```
def predict(theta,e,xtest):
    a=np.dot(xtest,theta)
    a=(a-a.mean())/a.std()
    return sigmoid(a)>=e
```

```
def accuracy(ypredicted,yactual):
    a=(yactual==ypredicted)
    return (a.sum()/a.shape[0])*100
```
